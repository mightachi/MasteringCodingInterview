# 3. Answer framework for Agoda backend/platform round

Use this structure for almost every design question.

## Step 1 — Clarify the problem
Ask:
- What is the user flow?
- What is the scale? QPS, data size, latency SLA, freshness SLA?
- Is this read-heavy or write-heavy?
- Are we optimizing for freshness, cost, availability, or consistency?
- Are external suppliers involved?
- What failure modes matter most?

## Step 2 — Define functional requirements
State them clearly:
- Search / booking / pricing / feature generation / supplier sync
- deduplication
- latest state view
- analytics / auditing
- operational controls

## Step 3 — Define non-functional requirements
Always mention:
- scalability
- availability
- latency
- durability
- correctness
- cost
- observability
- security
- extensibility

## Step 4 — Propose the high-level architecture
Give the main boxes first:
- clients
- API gateway / load balancer
- stateless services
- cache
- OLTP store / search index / object storage / warehouse
- Kafka / queue / stream processors
- feature or model services if relevant
- monitoring / alerting

## Step 5 — Walk the critical flows
Typically:
- write path / ingestion path
- read path / query path
- async refresh path
- failure handling path
- backfill / replay path

## Step 6 — Talk through trade-offs
This is where Agoda seems to care a lot.
Examples:
- cache freshness vs supplier API cost
- strong consistency vs low latency
- batch recomputation vs incremental updates
- push vs pull from suppliers
- SQL vs Elasticsearch
- Redis vs DB vs object store
- synchronous vs async enrichment
- precompute vs compute-on-read

## Step 7 — Add production-readiness
Mention:
- idempotency keys
- retries with backoff
- circuit breakers
- rate limiting
- dead-letter queue
- schema versioning
- metrics / logs / traces
- SLOs and alerts
- canary / shadow rollout

## Step 8 — Scale-up discussion
End with:
- partitioning / sharding
- hot key mitigation
- fan-out reduction
- multi-region considerations
- cost controls
- operational dashboards

# 5. Detailed answer playbook — supplier / inventory / booking systems

## Question: Design a flight or hotel inventory aggregation system

### A strong answer structure

### Problem
We need to serve search results quickly while combining:
- internal inventory / metadata
- external supplier inventory
- frequent updates
- duplicate offers from suppliers
- cost constraints if supplier APIs are metered

### Core requirements
- low latency for user search
- acceptable freshness SLA
- deduplicate same flight/hotel-room offer from multiple suppliers
- choose best offer (often cheapest valid one)
- degrade gracefully when supplier APIs are down
- keep auditability and observability

### Good high-level design
1. **Search API**
2. **Query planner / aggregator**
3. **Inventory cache**
4. **Supplier connector service**
5. **async refresh workers**
6. **canonicalization + dedup engine**
7. **pricing/offer ranking**
8. **search index / serving store**
9. **event bus + monitoring**

### Key idea
Do **not** call every supplier synchronously for every user request.
Use a hybrid model:
- serve from **fresh-enough cache**
- selectively refresh in background
- do synchronous fetch only for misses/high-value requests/critical stale cases

### Important trade-offs
- lower latency → more cache usage
- better freshness → more supplier API cost
- more availability → tolerate slightly stale data
- higher correctness → stronger dedup and reconciliation logic

### Smart refresh policy
Refresh based on:
- route / destination popularity
- time-to-departure or booking horizon
- supplier reliability score
- volatility of prices
- cache hit/miss trends
- revenue impact

### Dedup approach
Create canonical key using:
- supplier-independent identifiers if available
- normalized route/hotel + room + checkin/checkout + cancellation/refund attributes
- fuzzy reconciliation where IDs differ

### Failure handling
- timeout budgets per supplier
- partial results allowed
- circuit breakers
- stale fallback
- supplier health scores
- async revalidation

### Metrics
- p50/p95 latency
- freshness lag
- supplier API cost
- cache hit ratio
- stale-result ratio
- dedup collision rate
- timeout / error rate by supplier
- booking conversion impact

### Strong closing line
“I’d optimize for cached fast reads with selective refresh, because in travel search the business usually needs a balance of latency, freshness, and supplier cost rather than strict real-time reads from every upstream.”

# 10. How to sound senior/staff level

## What strong candidates do
- clarify requirements before drawing boxes
- make trade-offs explicit
- prioritize bottlenecks
- discuss failure modes early
- tie design choices to business constraints
- mention observability and operational readiness
- adapt the design as scale grows
- speak in terms of ownership and decision quality

## What weak candidates do
- jump straight to tools
- say “use Kafka, Redis, Mongo” without saying why
- ignore cost/freshness trade-offs
- ignore data quality and replay
- ignore rollout/rollback
- give a perfect architecture with no operational details

## Senior/staff phrasing you can use
- “I’d first optimize the hot read path because that’s the customer-facing SLA.”
- “I’d separate the serving path from the recomputation path to control tail latency.”
- “At this scale, data quality and idempotency become first-class concerns.”
- “I’d pick eventual consistency here because availability and latency matter more than fully synchronized reads.”
- “I’d instrument freshness lag and supplier health because otherwise we’re blind to silent degradation.”
- “I’d start simple, but I know how I’d evolve the design at 10x traffic.”
